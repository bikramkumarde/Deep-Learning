```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Reading data from dataset

```python
data = pd. read_csv('food_truck_data.txt', header = None)
```

Plotting the data using scatter plot and converting the data into individual columns and then converting it into a numpy array

```python
X = data.iloc[:, 0].values
Y = data.iloc[:, 1].values
plt.xlabel('Population')
plt.ylabel('Profit')
plt.scatter(X, Y)
plt.show()
```

Performing the linear regression: Here we perform the gradient descent algorithm. m and c are the gradient descent parameters. We calculate the parameters and update them simultaneously in each step. The update is done by calculating partial differential of the cost function with respect to each of the parameters. The cost function is stored in the cost_func array. The J_arr array is created to help in plotting the cost functions. It simply stores the epoch number. In the end, I have printed the final linear regression equation which shows the final calculated parameters. The epochs and the learning rate L can be set by the user.

```python
m = 0
c = 0
L = 0.001  # The learning Rate
epochs = 1000   #Number of times to perform linear regression
cost_func = []
cost_sum = 0
J_arr = []
count = 0
n = float(len(X)) # Number of elements in X
# Performing Gradient Descent
for j in range (epochs):
    for i in range(len(X)): 
        Y_pred = m*X[i] + c  # The current predicted value of Y
        D_m = (-2/n) * (X[i] * (Y[i] - Y_pred))  # Derivative wrt m
        D_c = (-2/n) * (Y[i] - Y_pred)  # Derivative wrt c
        m = m - L * D_m  # Update m
        c = c - L * D_c  # Update c
        cost_sum += np.square((Y[i]- Y_pred))
    cost = cost_sum/(n*2)
    cost_func.append(cost)
    cost_sum = 0
    J_arr.append(count)
    count+= 1
        
print('-----------Equation--------------')    
print ('Y =',m,'*X +', c)
```

```python
predicted_arr = []   #Storing all the predicted values for future use
for i in range(len(X)):
    Y_pred = m*X[i] + c
    predicted_arr.append(Y_pred)
```

Plotting the data using scatter plot and the predicted regression line is shown in red

```python
plt.title('Linear Regression using gradient descent')
plt.xlabel('Population')
plt.ylabel('Profit')
plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(predicted_arr), max(predicted_arr)], color='red')  # regression line
plt.show()
```

Plotting the cost function as a function of the gradient descent iterations. We can see that the cost function decreases very fast initially but after a certain number of iterations the rate of decrease is much less which indicates that our linear regression line is gradually becomming closer to the actual value, hence the error is reducing asymptotically.

```python
plt.title('Cost Function Graph')
plt.xlabel('Epoch')
plt.ylabel('Cost Function')
plt.plot(J_arr, cost_func)
plt.show()
```
